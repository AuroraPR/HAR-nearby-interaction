```python
import math


ini_date = '12/08/22 13:00:00'
end_date = '13/08/22 16:45:00'

time_step=15

smooth=True

def filter0(x,y):
    return x,y

def distance(x1,y1,x2,y2):
    return math.sqrt(((x2 - x1) ** 2) + (y2 - y1) ** 2)

def aggretation_function(xsA,ysA,xsB,ysB):
    if(len(xsA)==0):
        return -1
    if(len(xsB)==0):
        return -1    
    d=0
    #print(len(xsA),len(xsB))
    mind=float("inf")
    for i in range(0,len(xsA)):
        for j in range(0,len(xsB)):
            mind=min(mind,distance(xsA[i],ysA[i],xsB[j],ysB[j]))

    return mind

def uncertain(xsA,ysA,xsB,ysB):
    w=float(len(xsA))/5.0
    if(w>=1):
        w=1.0
    return w

tags=["1", "2"]

filter_function={}
filter_function["1"]=filter0
filter_function["2"]=filter0


colors={}
colors["1"]="red"
colors["2"]="blue"

def repaintXY(x,y):
    return int((x+500)/13.5), int((y+6300)/13.5)
```

```python
def Qd(x,l,u):
    if(x<=l):
        return 1
    if(x>=u):
        return 0
    return 1.0-(x-l)/(u-l)
```

```python
import numpy as np
from pandas import read_csv    

import time

from datetime import datetime

def day_time(ti):
    return int((int)(ti/(60*60*24)))



t0 = (int)(datetime.strptime(ini_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", t0, "day:",day_time(t0))

tN = (int)(datetime.strptime(end_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", tN, "day:",day_time(tN))

def relT(ti):
    return (int)((ti-t0)/time_step)

nT=(int)(round((tN-t0-1)/time_step))
print(nT)
```

```python
dt={}
ts={}
txy={}
for tag in tags:
    txy[tag]=[]
    for d in range(day_time(t0),day_time(tN)+1):
        print("Loading day ",d)
        try:
            series=read_csv("./data/uwb/"+str(d)+"."+tag+".tsv", sep="\t",parse_dates=True,header=None)
            lastT=10000
            lastX=0
            lastY=0
            for vt in series.values[:,[0,2,3]]:
                x,y=filter_function[tag](vt[1],vt[2])
                t=relT(vt[0])
                
                #print("vt:",vt)
                print("incT:",t-lastT)
                
                if smooth:
                    if(t-lastT == 0):
                        x=lastX*0.75+x*0.25
                        y=lastY*0.75+y*0.25

                    if(t-lastT == 1):
                        x=lastX*0.5+x*0.5
                        y=lastY*0.5+y*0.5

                    if(t-lastT == 2):
                        x=lastX*0.25+x*0.75
                        y=lastY*0.25+y*0.75

                    lastT=t
                    lastX=x
                    lastY=y
                
                txy[tag].append([t,x,y])
                
        except Exception as e: 
            print(e)
            print(" Not data for tag ",tag, " in day ",d, "./data/uwb/"+str(d)+"."+tag+".tsv", "vs:19216.1")
    txy[tag]=np.array(txy[tag])
    print(txy[tag].shape)
    print(txy[tag])
    ts[tag]=txy[tag][:,0]
    




        
```

```python
I=np.zeros((len(tags),len(tags), nT))
print(I.shape)
for iA,tagA in enumerate(tags):
    
    for iB,tagB in enumerate(tags):
        if tagA==tagB:
            continue
            
        for t in range(0,nT):
           
            ixA=np.where( (ts[tagA]==t))[0]
            ixB=np.where( (ts[tagB]==t))[0]
            if(len(ixA)>0 and len(ixB)):
                xsA,ysA=txy[tagA][:,1][ixA],txy[tagA][:,2][ixA]
                xsB,ysB=txy[tagB][:,1][ixB],txy[tagB][:,2][ixB]
                r=aggretation_function(xsA,ysA,xsB,ysB)
                w=uncertain(xsA,ysA,xsB,ysB)
                print(iA,iB,t,xsA,ysA,xsB,ysB, "R:",r)
                I[iA][iB][t]=Qd(r,500,1000)
            else:
                I[iA][iB][t]=-1
```

```python
import pandas as pd
def prinUserUserLocation(I):
    df = pd.DataFrame()
    ts=[]
    for t in range(t0,tN,time_step):
        ts.append(datetime.fromtimestamp( t))

    df['Time']=ts
    ls=[]
    for iA,tagA in enumerate(tags):
        for iB,tagB in enumerate(tags):
            if tagA==tagB:
                continue
            df[tagA+"-"+tagB]=I[iA][iB]
            ls.append(tagA+"-"+tagB)
    df.plot(x="Time",y=ls,subplots=True, legend=True, ylim=(0,1), figsize=(10,10))    
    
prinUserUserLocation(I)
```

```python
from tkinter import *
from PIL import ImageTk, Image
from pandas import read_csv
import time
from threading import Thread
import threading

speed=0.05

tt=t0

con = threading.Condition()

images = []  # to hold the newly created image

def create_rectangle(x1, y1, x2, y2, **kwargs):
    if 'alpha' in kwargs:
        alpha = int(kwargs.pop('alpha') * 255)
        fill = kwargs.pop('fill')
        fill = win.winfo_rgb(fill) + (alpha,)
        image = Image.new('RGBA', (x2-x1, y2-y1), fill)
        images.append(ImageTk.PhotoImage(image))
        canvas.create_image(x1, y1, image=images[-1], anchor='nw')
    canvas.create_rectangle(x1, y1, x2, y2, **kwargs)
    
def create_text(canvas, text): #center coordinates, radius
    canvas.create_rectangle(
    50, 5, 200, 20,
    outline="#fb0",
    fill="#fb0")
    canvas.create_text(100,10,fill="darkblue",text=text)

def create_big_text(canvas, text): #center coordinates, radius
    canvas.create_rectangle(
    50, 35, 200, 55,
    outline="#fb0",
    fill="#fb0")
    canvas.create_text(150,30,fill="darkblue",text=text)    

def create_circle(x, y, r, canvas, color): #center coordinates, radius
    x0 = x - r
    y0 = y - r
    x1 = x + r
    y1 = y + r
    canvas.create_oval(x0, y0, x1, y1, fill=color)

def create_bb(x1, y1, x2, y2, canvas, color, text, alpha=.5): #center coordinates, radius
    create_rectangle(
    x1, y1, x2, y2,    outline=color,
    fill=color, alpha=alpha)
    canvas.create_text((x1+x2)/2,(y1+y2)/2,fill=color,text=text)

def paintUserLocation(tt):
    t=relT(tt)
    good=True
    for tag in tags:
        if( 'None' in user_location[tag][t]):
            continue

        user_location[tag][t]=user_location[tag][t].astype(np.float)
        time.sleep(speed)
        x1,y1,x2,y2,w=user_location[tag][t][0],user_location[tag][t][1],user_location[tag][t][2],user_location[tag][t][3],user_location[tag][t][4]
        print(tag,datetime.fromtimestamp(tt).strftime("%Y-%m-%d %H:%M:%S"), " x1,y1,x2,y2:",x1,y1,x2,y2)
        x1,y1=repaintXY(x1,y1)
        x2,y2=repaintXY(x2,y2)
        if(x1<0 or y1<0):
            good=False
        print("\t",tag,datetime.fromtimestamp(tt).strftime("%Y-%m-%d %H:%M:%S"), " x1,y1,x2,y2:",x1,y1,x2,y2)
        create_bb(x1,y1, x2,y2, canvas, colors[tag], tag+"("+str(t)+")",w*0.5)
        for i in np.where( (ts[tag]==t))[0]:
            xs=txy[tag][:,1][i]
            ys=txy[tag][:,2][i]
            print("\t",i,xs,ys)
            xs,ys=repaintXY(xs,ys)
            create_circle(xs,ys,10, canvas, colors[tag])
        
    return good
        

        
def paintUserInteraction(tt):
    t=relT(tt)
    canvas.create_image(500/2, 320/2, image=img)
    create_text(canvas, datetime.fromtimestamp(tt).strftime("%Y-%m-%d %H:%M:%S"))
    paintUserLocation(tt)

    con.acquire()
    con.wait()
    con.release()
            

def paint(tt):
    t=relT(tt)
    create_text(canvas, datetime.fromtimestamp(tt).strftime("%Y-%m-%d %H:%M:%S"))

    return paintUserLocation(tt)

        
def print0():
    
    for tt in range(t0,tN,time_step):
        t=relT(tt)
        good=paint(tt)
            
        if(t%10==0):
            canvas.create_image(500/2, 320/2, image=img)
            
            
                
            
        for iA,tagA in enumerate(tags):
            for iB,tagB in enumerate(tags):
                if tagA==tagB:
                    continue
                if(I[iA][iB][t]>0 and good):
                    print("INTERACTION at ",t,I[iA][iB][t])
                    paintUserInteraction(tt)
            

def callback(event):
    print( "clicked at", event.x, event.y)
    con.acquire()
    con.notify()
    con.release()

def past(e):
        print( "past at")
        global tt
        tt=tt-time_step
        t=relT(tt)
        paint(tt)
def future(e):
        print( "future at")
        global tt
        tt=tt+time_step
        t=relT(tt)
        paint(tt)
        
#Create an instance of tkinter window
win = Tk()

# Define the geometry of the window
#win.geometry("1010x612")

#frame = Frame(win, width=600, height=400)
#frame.pack()
#frame.place(anchor='center', relx=0.5, rely=0.5)

# Create an object of tkinter ImageTk
img = ImageTk.PhotoImage(Image.open("cocina.png"))

# Create a Label Widget to display the text or Image
#label = Label(frame, image = img)

#label.pack()
#label.bind('<Button 1>', motion)

canvas = Canvas(win, height=320, width=500)
canvas.create_image(500/2, 320/2, image=img)
canvas.bind("<Button-1>", callback)
win.bind("<Left>", past)
win.bind("<Right>", future)
#print0() 
canvas.pack()
print("pre loop")



thread = Thread(target = print0, args = ())
thread.setDaemon(True)
thread.start()

    
win.mainloop()


    


```
