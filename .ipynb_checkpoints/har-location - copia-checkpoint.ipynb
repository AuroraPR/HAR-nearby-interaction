```python


ini_date = '12/08/22 13:00:00'
end_date = '13/08/22 17:00:00'

time_step=60


def filter0(x,y):
    return x,y

def agg0(xs,ys):
    if(len(xs)==0):
        return None
    return np.mean(xs), np.mean(ys)


tags=["1", "2"]

filter_function={}
filter_function["1"]=filter0
filter_function["2"]=filter0


aggretation_function={}
aggretation_function["1"]=agg0
aggretation_function["2"]=agg0


colors={}
colors["1"]="red"
colors["2"]="blue"

```

```python
import numpy as np
from pandas import read_csv    

import time

from datetime import datetime

def day_time(ti):
    return int((int)(ti/(60*60*24)))



t0 = (int)(datetime.strptime(ini_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", t0, "day:",day_time(t0))

tN = (int)(datetime.strptime(end_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", tN, "day:",day_time(tN))

def relT(ti):
    return (int)((ti-t0)/time_step)
```

```python
dt={}
for sensor in tags:
    txy=[]
    for d in range(day_time(t0),day_time(tN)+1):
        print("Loading day ",d)
        try:
            series=read_csv("./data/uwb/"+str(d)+"."+sensor+".tsv", sep="\t",parse_dates=True,header=None)
            for vt in series.values[:,[0,2,3]]:
                #print("vt:",vt)
                t=relT(vt[0])
                x,y=filter_function[sensor](vt[1],vt[2])
                if(x):
                    txy.append([t,x,y])
                
        except Exception as e: 
            print(e)
            print(" Not data for sensor ",sensor, " in day ",d, "./data/uwb/"+str(d)+"."+sensor+".tsv", "vs:19216.1")
    txy=np.array(txy)
    print(txy.shape)
    print(txy)
    ts=txy[:,0]
    
    #f = open("./time-data/"+sensor+".tsv", "w")
    vs=[]
    for t in range(t0,tN,time_step):
        ti=relT(t)
        i=np.where( (ts==ti))[0]
        if(len(i)>0):
            print(i, " dt[i]:",txy[:,1][i])
            agg_x,agg_y=aggretation_function[sensor](txy[:,1][i],txy[:,2][i])

            print(ti," agg_x:",agg_x," agg_x:",agg_y)
            vs.append([agg_x,agg_y])
        else:
            vs.append([None,None])
      #  f.write(str(ti)+"\t"+str(agg_x)++"\t"+str(agg_y)+"\t"+datetime.fromtimestamp( t).strftime("%Y-%m-%d %H:%M:%S")+"\n")
    dt[sensor]=  np.array(vs)  
    #f.close()



        
```
