```python


ini_date = '12/08/22 13:00:00'
end_date = '13/08/22 16:45:00'

time_step=15


binary_sensors=["clutery", "dishwasher","fridge","microwave"]

#["clutery", "cupboard", "dishwasher","drawer_vitro","fridge","microwave", "oven", "vibration_cupboard", "vibration_tap_kitchen", "washing_machine"]
```

```python
import numpy as np
from pandas import read_csv    

import time

from datetime import datetime

def day_time(ti):
    return int((int)(ti/(60*60*24)))



t0 = (int)(datetime.strptime(ini_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", t0, "day:",day_time(t0))

tN = (int)(datetime.strptime(end_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", tN, "day:",day_time(tN))

def relT(ti):
    return (int)((ti-t0)/time_step)
```

```python
import pandas as pd

df = pd.DataFrame()

ts=[]
for t in range(t0,tN,time_step):
    ts.append(datetime.fromtimestamp( t))

df['Time']=ts    
    
```

```python

for sensor in binary_sensors:
    lastV=-1
    dt=[]
    cS=0
    for d in range(day_time(t0),day_time(tN)+1):
        print("Loading day ",d)
        try:
            series=read_csv("./data/binary/"+str(d)+"."+sensor+".tsv", sep="\t",parse_dates=True,header=None)
            for vt in series.values[:,[0,2]]:
                cS=cS+1
                if(lastV != vt[1]):
                    dt.append([relT(vt[0]),1])
                else:
                    #print("ignoring ",vt[0])
                    dt.append([relT(vt[0]),1])
                lastV=vt[1]
        except:
            print(" Not data for sensor ",sensor, " in day ",d)
    
    print("#Sensor:",sensor, " #:",cS)
    dt=np.array(dt)
    #print(dt.shape)
    #print(dt)
    ts=dt[:,0]
    f = open("./time-data/"+sensor+".tsv", "w")
    vs=[]
    cS=0
    for t in range(t0,tN,time_step):
        ti=relT(t)
        i=np.where( (ts==ti))[0]
        if(len(i)!=0):
            cS=cS+1
            vs.append(1)
            f.write(str(ti)+"\t1\t"+datetime.fromtimestamp( t).strftime("%Y-%m-%d %H:%M:%S")+"\n")
        else:
            vs.append(0)
            f.write(str(ti)+"\t0\t"+datetime.fromtimestamp( t).strftime("%Y-%m-%d %H:%M:%S")+"\n")
    f.close()
    df[sensor]=vs
    print("#Sensor:",sensor, " #:",cS)


        
```

```python
df.plot(x="Time",y=binary_sensors,subplots=True, legend=True, figsize=(6,15))
```
