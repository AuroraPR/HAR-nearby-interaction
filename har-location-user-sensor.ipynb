```python


ini_date = '12/08/22 13:00:00'
end_date = '13/08/22 16:45:00'

time_step=15

smooth=True

def filter0(x,y):
    return x,y

def agg0(xs,ys):
    if(len(xs)==0):
        return None, None, None, None,None
    return np.min(xs), np.min(ys), np.max(xs), np.max(ys), np.mean(xs), np.mean(ys)


def uncertain(xs,ys):
    w=float(len(xs))/5.0
    if(w>=1):
        w=1.0
    return w

tags=["1", "2"]

filter_function={}
filter_function["1"]=filter0
filter_function["2"]=filter0


aggretation_function={}
aggretation_function["1"]=agg0
aggretation_function["2"]=agg0


colors={}
colors["1"]="red"
colors["2"]="blue"

```

```python
import numpy as np
from pandas import read_csv    

import time

from datetime import datetime

def day_time(ti):
    return int((int)(ti/(60*60*24)))



t0 = (int)(datetime.strptime(ini_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", t0, "day:",day_time(t0))

tN = (int)(datetime.strptime(end_date, '%d/%m/%y %H:%M:%S').timestamp())
print ("Init date is", tN, "day:",day_time(tN))

def relT(ti):
    return (int)((ti-t0)/time_step)
```

```python
dt={}
ts={}
txy={}

for tag in tags:
    txy[tag]=[]
    for d in range(day_time(t0),day_time(tN)+1):
        print("Loading day ",d)
        try:
            series=read_csv("./data/uwb/"+str(d)+"."+tag+".tsv", sep="\t",parse_dates=True,header=None)
            lastT=10000
            lastX=0
            lastY=0
            for vt in series.values[:,[0,2,3]]:

                x,y=filter_function[tag](vt[1],vt[2])
                t=relT(vt[0])
                
                #print("vt:",vt)
                #1818print("incT:",t-lastT)
                
                if smooth:
                    if(t-lastT == 0):
                        x=lastX*0.75+x*0.25
                        y=lastY*0.75+y*0.25

                    if(t-lastT == 1):
                        x=lastX*0.5+x*0.51818
                        y=lastY*0.5+y*0.5

                    if(t-lastT == 2):
                        x=lastX*0.25+x*0.75
                        y=lastY*0.25+y*0.75

                    lastT=t
                    lastX=x
                    lastY=y
                
                txy[tag].append([t,x,y])
                
        except Exception as e: 
            print(e)
            print(" Not data for tag ",tag, " in day ",d, "./data/uwb/"+str(d)+"."+tag+".tsv", "vs:19216.1")
    txy[tag]=np.array(txy[tag])
    
    print("#txy[tag]:",txy[tag].size)
    
    print(txy[tag].shape)
    print(txy[tag])
    ts[tag]=txy[tag][:,0]
    
    f = open("./time-data/"+tag+".tsv", "w")
    vs=[]
    cL=0
    for t in range(t0,tN,time_step):
        ti=relT(t)
        i=np.where( (ts[tag]==ti))[0]
        print(tag, ti)
        if(len(i)>0):
            cL=cL+1
            #print(datetime.fromtimestamp( t).strftime("%Y-%m-%d %H:%M:%S"), " xs:",txy[sensor][:,1][i], " ys:",txy[sensor][:,2][i])
            xs,ys=txy[tag][:,1][i],txy[tag][:,2][i]
            x1,y1,x2,y2,x3,y3=aggretation_function[tag](xs,ys)
            w=uncertain(xs,ys)
            #print(ti," x1,y1,x2,y2,x3,y3:",x1,y1,x2,y2,x3,y3)
            vs.append([x1,y1,x2,y2,x3,y3])
            f.write(str(ti)+"\t"+str(x1)+"\t"+str(y1)+"\t"+str(x2)+"\t"+str(y2)+"\t"+str(w)+"\t"+datetime.fromtimestamp( t).strftime("%Y-%m-%d %H:%M:%S")+"\n")
        else:
            vs.append([None,None,None,None,None,None])
            f.write(str(ti)+"\t"+str(None)+"\t"+str(None)+"\t"+str(None)+"\t"+str(None)+"\t"+str(None)+"\t"+datetime.fromtimestamp( t).strftime("%Y-%m-%d %H:%M:%S")+"\n")
    dt[tag]=  np.array(vs)
    print("#Tag:",tag, " #:",cL)
    f.close()



        
```
